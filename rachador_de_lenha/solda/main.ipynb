# Cálculos de Solda

O material base da lâmina é o Aço AISI 420. Informações importantes do mesmo podem ser observadas na figura abaixo:

```python
Image('aisi420.png')
```

A tabela abaixo mostra a lista de eletrodos para realizar a solda.

A recomendação básica para a seleção dos eletrodos é que a resistência deles (tensão máxima) seja
menor ou igual à resistência do material base.

Escolheu-se o E90xx para atender a esse requisito

```python
Image('eletrodo.png')
```

```python
Exx = 620 # N/mm²
```

$$\tau_{adm} = 0.30Exx$$

```python
tau_adm = 0.30*Exx
display(Math('\\tau_{adm} = %s MPa'%tau_adm))
```

```python
F = 15000 # Requisito do projeto
P = F * 9.81 # Deixando em N
A_cis = sy.symbols('A_cis')
```

```python
eq = sy.Eq(tau_adm, P/A_cis)
eq
```

```python
area_cis = sy.solve(eq, A_cis)[0]
display(Math('A_{cis} = %s mm²'%round(area_cis,2)))
```

```python
L = 150  #mm  #é o comprimento da lâmina
```

A lâmina será soldada em T em ambos os lados da chapa.

```python
t = area_cis / (2*L)
# Em que t é a dimensão da garganta

display(Math(f't = {t:.2f} mm'))
```

Convertendo a dimensão da garganta, t, para a dimensão das pernas, w

```python
w = t / math.cos(math.radians(45))
display(Math(f'w = {w:.2f} mm'))
```

Verificando a recomendação mínima do tamanho do cordão:

```python
Image('recomendação.png')
```

Verificando falhas na área de fusão

```python
A_fusao = 2*L*w

tau_xy = P / A_fusao
display(Math('\\tau_{xy} = %s MPa'%round(tau_xy,2)))
```
