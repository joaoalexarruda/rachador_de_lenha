# Cálculo do dimensionamento da lâmina

Para realizar esse cálculo, precisa-se do material utilizado na lâmina, área da lâmina e sua geometria, força, resistência ao cisalhamento da madeira.

```python
Image('tab_cisalhamento_madeira.jpeg')
```

```python
sigma = 9.93*10**6   # Pa
A = sy.symbols('A')
```

```python
F = 15000    # kgf
```

```python
sigma = sigma / 9.81
# Dividiu-se pela aceleração da gravidade para ajustar as unidades.
```

```python
eq = sy.Eq(sigma, F/A)
eq
```

```python
area = sy.solve(eq, A)[0]
display(Math(f'A = {area:.5f}m² = {area*10**6:.2f}mm²'))
```
